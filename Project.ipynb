```python
#Code to fetch top news hashtags
company=data["Company"].tolist()
handle_list=[]
for i in company:
    klist=[]
    company_data=df_dict[i]
    for i in company_data['handles']:
        if len(i)>0:
            for element in i:
                if element not in klist:
                    klist.append(element)                 
    handle_list.append(klist)
flat_list = [item for sublist in handle_list for item in sublist]
dict = {} 
for w in flat_list:
    dict[w] = dict.get(w, 0) + 1

handle_freq = []
for key, value in dict.items():
    handle_freq.append((value, key))
handle_freq.sort(reverse=True)
```

# Fetch the URl  for news websites using news handles filtered above as the keywords in the urls tweeted

```python
import urllib.request
handle_dict={}
company=data["Company"].tolist()
for i in company:
    url_list=[]
    valid_url_list=[]
    company_data=df_dict[i]
    for tweet in company_data["text"]:
        urls = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', tweet)
        for url in urls:
         
            for handle in final_handle_list:
                if(len(handle)!=0):
                    if handle.lower() in url.lower():
                        if(len(handle)!=0):
                            valid_url_list.append(url)
    handle_dict[i]=valid_url_list

        
```

# Get common hashtags according to category

```python
topic_list=data["Topic"].dropna().unique().tolist()
hashtag_dict={}
len_dict={}
for topic in topic_list:
    topic_data=data.loc[data["Topic"]==topic]
    len_dict[topic]=len(topic_data)
    company_list=topic_data["Company"].unique().tolist()
    hashtag_list=[]
    for company in company_list:
        klist=[]      
        frame=df_dict[company]
        for i in frame['hashtags']:
            if len(i)>0:
                for element in i:
                    if element not in klist:
                        klist.append(element)
        hashtag_list.append(klist)
        
    flat_list = [item for sublist in hashtag_list for item in sublist]
    dict = {} 
    for w in flat_list:
        dict[w] = dict.get(w, 0) + 1

    word_freq = []
    for key, value in dict.items():
        word_freq.append((value, key))
    word_freq.sort(reverse=True)
    hashtag_dict[topic]=word_freq
            
```

```python
with open("hashtag_dict.pkl", "wb") as f:
    pickle.dump(hashtag_dict, f) 
with open("len_dict.pkl", "wb") as f:
    pickle.dump(len_dict, f) 
```

# Filtering top hashtags for Racial and Databrach categories that are shown in graphs

```python
classification_hashtags={}
for keys in word_repositories.keys():
    words_to_plot=[]
    for hashtg_tuple in hashtag_dict[keys]:
        for i in word_repositories[keys]:
            if(i in hashtg_tuple[1] and hashtg_tuple[0]>2):
                words_to_plot.append(hashtg_tuple)
                words_to_plot=list(set(words_to_plot))
                words_to_plot.sort(reverse=True)
    values, labels = zip(*words_to_plot)
    classification_hashtags[keys]=list(labels)
    plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.bar(range(0,len(values)),values,tick_label=labels,color="red")
    plt.xlabel('Hashtags')
    plt.xticks(rotation=90,color="Black")
    plt.ylabel('Count')
    plt.title('Hashtags representing racial category amongst total of '+str(len_dict['Racial'])+" companies")
    plt.show()  
with open("classification_hashtags.pkl", "wb") as f:
    pickle.dump(classification_hashtags, f)
```
